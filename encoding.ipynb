```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder


df=pd.read_csv("sample_data.csv")

df.head()

```

```python
df_label = df.copy()
le = LabelEncoder()

df_label["Label_gender"] = le.fit_transform(df_label["Gender"])
df_label["Label_Passed"] = le.fit_transform(df_label["Passed"])

print(df_label[["Gender","Passed","Label_gender","Label_Passed"]])
```
