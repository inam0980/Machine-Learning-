```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load
df = pd.read_csv("/Users/inamkhan/Desktop/MAIN/Machine Learning/Machine Learning FUll/student1.csv")

# New df (sirf zaroori columns)
df_ml = df[["Grade", "Study_Hours_per_Week", "Final_Score"]]

# Drop trick
X = df_ml.drop("Final_Score", axis=1)
y = df_ml["Final_Score"]

# Train-Test Split (FIRST)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Encoding (AFTER split)
X_train = pd.get_dummies(X_train, columns=["Grade"], drop_first=True)
X_test  = pd.get_dummies(X_test, columns=["Grade"], drop_first=True)

# Align columns
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)

# Scaling (sirf numeric)
sc = StandardScaler()
X_train[["Study_Hours_per_Week"]] = sc.fit_transform(
    X_train[["Study_Hours_per_Week"]]
)
X_test[["Study_Hours_per_Week"]] = sc.transform(
    X_test[["Study_Hours_per_Week"]]
)

# Model (Regression)
model = LinearRegression()
model.fit(X_train, y_train)

y_pred=model.predict(X_test)

# Example prediction
print(model.predict(X_test[:1])[0])

print("R2 Score: ",r2_score(y_test,y_pred))
```
