```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

dfl=pd.read_csv("/Users/inamkhan/Desktop/MAIN/Machine Learning/Machine Learning FUll/student1.csv")
df=dfl[["Gender","Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg","Placement_Status"]]


X=df[["Gender","Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg"]]
y=df["Placement_Status"]

X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3,random_state=42)

le=LabelEncoder()
X_train['Gender']=le.fit_transform(X_train['Gender'])
X_test['Gender']=le.transform(X_test["Gender"])

sc=StandardScaler()

X_train[["Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg"]]=sc.fit_transform(X_train[["Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg"]])
X_test[["Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg"]]=sc.transform(X_test[["Attendance_%","Study_Hours_per_Week","Final_Score","Assignments_Avg"]])

model=KNeighborsClassifier(n_neighbors=3)

model.fit(X_train,y_train)

y_pred=model.predict(X_test)
print("Accuracy: ",accuracy_score(y_test,y_pred))
print(y_pred)
```
