```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score , precision_score,f1_score,recall_score
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

dfl=pd.read_csv("/Users/inamkhan/Desktop/MAIN/Machine Learning/Machine Learning FUll/student1.csv")
df=dfl[["Department","Attendance_%","Study_Hours_per_Week","Midterm_Score","Grade"]]

X=df[["Department","Attendance_%","Study_Hours_per_Week","Midterm_Score"]]
y=df["Grade"]
X_train,X_test,y_train,y_test= train_test_split( X, y, test_size=0.3, random_state=42)

X_train=pd.get_dummies(X_train,columns=["Department"], drop_first=True)

X_test=pd.get_dummies(X_test,columns=["Department"], drop_first=True)

le=LabelEncoder()
y_train=le.fit_transform(y_train)

y_test=le.transform(y_test)

sc=StandardScaler()

X_train[['Attendance_%','Study_Hours_per_Week','Midterm_Score']]=sc.fit_transform(X_train[['Attendance_%','Study_Hours_per_Week','Midterm_Score']])
X_test[['Attendance_%','Study_Hours_per_Week','Midterm_Score']]=sc.transform(X_test[['Attendance_%','Study_Hours_per_Week','Midterm_Score']])

model=SVC(kernel="rbf")
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

print(le.inverse_transform(y_pred))


print("Accuracy Score:- ",accuracy_score(y_test,y_pred))
print("Precision:-", precision_score(y_test, y_pred, average='weighted'))
print("Recall:-", recall_score(y_test, y_pred, average='weighted'))
print("F1-score:-", f1_score(y_test, y_pred, average='weighted'))

```
