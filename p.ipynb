```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("sample_data.csv")
df.head()
```

```python
le=LabelEncoder()

df["Gender"]=le.fit_transform(df["Gender"])

df.head()
```

```python
df = pd.get_dummies(df, columns=["City"])

df.head()
```

```python
df["City_Bangalore"]=le.fit_transform(df["City_Bangalore"])
df["City_Chennai"]=le.fit_transform(df["City_Chennai"])
df["City_Mumbai"]=le.fit_transform(df["City_Mumbai"])
df["City_Delhi"]=le.fit_transform(df["City_Delhi"])
df.head()
```
